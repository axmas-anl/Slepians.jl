```python
import numpy as np
from astropy.timeseries import LombScargle
import matplotlib.pyplot as plt

rand = np.random.default_rng(42)

t = 100 * rand.random(100)

y = np.ones(100)

plt.scatter(t,y)
plt.show()
```

```python
frequency, power = LombScargle(t, y, fit_mean = False).autopower()

frequency[np.argmax(power)]  
```

```python
freq = np.arange(0.8, 1.3, 0.1)

LombScargle(t, y).power(freq)  
```

```python
frequency2, power2 = LombScargle(t, y, normalization='psd', fit_mean = False).autopower()
```

```python
plt.semilogy(frequency, power)
plt.semilogy(frequency2, power2)
plt.show()
```
