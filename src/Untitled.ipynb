```python
import numpy as np
import astropy.convolution as scp
import matplotlib.pyplot as plt

#see https://docs.astropy.org/en/stable/api/astropy.convolution.convolve.html
```

```python
x = np.linspace(0, 1, 200)*2*np.pi*6
y = np.sin(x)

# Replace y with NANs at these indices
ind = np.where(y >= 0.8)
```

```python
z = np.copy(y)
z[ind] = np.nan

# Create a Gaussian kernel with which to convolve
x2 = np.linspace(-1, 1, 21)
ell = 0.001
gaussian = np.exp(-x2**2/ell)

plt.plot(x2,gaussian)
```

```python
w = scp.convolve(z, gaussian )

fig, ax = plt.subplots(1,1)
plt.plot(x, z)
plt.plot(x, w)
plt.plot(x, y)
plt.legend(["original", "punched", "interpolated, l = ell"])
```
